## The Case 

A data analyst was given the task by the head of the MIS office to look into the performance of a
remote procedure call (RPC) mechanism that was compared between two (2) mainframe
operating systems (OS) - UNIX and ARGUS. The performance metric was total elapsed
time, which was measured for various data sizes.

---

## Data Preparation

```python
import pandas as pd
```

```python
unix_rpc = {"Data Bytes":[64, 64, 64, 64, 234, 590, 846, 1060, 1082, 1088, 1088, 1088, 1088], 
            "Time": [26.4, 26.4, 26.4, 26.2, 33.8, 41.6, 50.0, 48.4, 49.0, 42.0, 41.8, 41.8, 42]}
argus_rpc = {"Data Bytes":[92, 92, 92, 92, 348, 604, 860, 1074, 1074, 1088, 1088, 1088, 1088], 
            "Time": [32.8, 34.2, 32.4, 34.4, 41.4, 51.2, 76.0, 80.8, 79.8, 58.6, 57.6, 59.8, 57.4]}
```

```python
df1 = pd.DataFrame(unix_rpc)
df2 = pd.DataFrame(argus_rpc) 
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1, df2])
```

```python
df['OS'] = ['Unix'] * len(df1) + ['Argus'] * len(df2)
```

```python
df
```

---

## Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(df, hue='OS', palette='magma')
# plt.title("Pairplot of Data Bytes vs. Time categorized by OS (Argus and Unix)") 
plt.suptitle("Pairplot of Data Size vs. Time colored by OS", y=1.02)
plt.show()
```
